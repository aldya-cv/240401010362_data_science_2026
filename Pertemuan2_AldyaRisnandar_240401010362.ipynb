```python
import pandas as pd
import numpy as np
```

```python
URL = "https://raw.githubusercontent.com/datasciencedojo/datasets/refs/heads/master/titanic.csv"
df = pd.read_csv(URL)
df.head()
```

```python
df.info()
df.describe()
```

```python
print("Q1: Berapa jumlah baris dan kolom pada dataset Titanic? Kolom apa saja yang tersedia?")

print("Baris:", df.shape)
print("Kolom:", df.columns.tolist())
```

```python
print("Q2: Kolom mana saja yang memiliki missing values? Berapa jumlah dan proporsinya (%) per kolom?")

missing = df.isnull().sum()
percent = (missing / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing Count': missing,
    'Persentase (%)': percent
})

missing_df
```

```python
print("Q3: Berapa persen penumpang yang selamat (Survived=1)? Gunakan operasi NumPy atau Pandas.")

survival_rate = df['Survived'].mean() * 100
print(f"Persentase selamat: {survival_rate:.2f}%")
```

```python
print("Q4: Filter penumpang wanita dari kelas 1 (Pclass=1). Berapa jumlahnya dan berapa rata- rata usianya?")

female_class1 = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]

jumlah = len(female_class1)
rata_umur = female_class1['Age'].mean()

print("Jumlah:", jumlah)
print("Rata-rata umur:", rata_umur)
```

```python
print("Q5: Gunakan groupby untuk menghitung tingkat keselamatan (mean Survived) per kelas penumpang. Kelas mana yang paling tinggi?")

survival_by_class = df.groupby('Pclass')['Survived'].mean()
print(survival_by_class)

print("Kelas paling tinggi = Kelas 1")
```

```python
print("Buat analisis tambahan berdasarkan variabel lain yang Anda anggap menarik. Sertakan interpretasi singkat.")

survival_by_gender = df.groupby('Sex')['Survived'].mean()
print(survival_by_gender)

print("Interpretasi: Mengambil variabel Sex dengan Survived, menghasilkan interpretasi perempuan memiliki probabilitas selamat jauh lebih tinggi dibanding laki-laki, mengindikasikan adanya prioritas evakuasi berbasis gender.")
```
